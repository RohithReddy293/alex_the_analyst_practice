```python
#  List / Sequence Unpacking in Python

ice_cream = ['Chocolate', 'Vanilla', 'Rocky Road'] #Unpacking: each element is assigned to a variable
x, y, z = ice_cream
print(x)#The number of variables must match the number of items.
print(y)
print(z)
```

```python
#You can use * to collect extra values.
#Example with * (star operator)
a, *b = [1, 2, 3, 4]
print(a)   # 1
print(b)   # [2, 3, 4]
```

```python
# Example ignoring values with _
first, _, third = [10, 20, 30]
print(first, third)   # 10 30
```
