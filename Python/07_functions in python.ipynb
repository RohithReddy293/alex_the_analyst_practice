A function is a reusable block of code that performs a specific task.
It:
takes inputs (called parameters or arguments),
does some processing,
and can return an output.
It helps you avoid repeating code and makes programs easier to read, test, and maintain.

```python
def first_function():
  print(" we did it")
```

```python
first_function()
```

```python
def number_squared(number):
    print(number**2)
```

```python
number_squared(5)
```

```python
def number_squared_custom(number,power):
    print(number**power)
```

```python
number_squared_custom(3,4)
```
