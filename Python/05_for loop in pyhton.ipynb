FOR LOOP
for variable in sequence:
    # code block
variable → a temporary name that takes each value from the sequence (list, tuple, string, range, etc.).
sequence → the collection or iterable you’re looping through.
The indented block runs once for every item.

```python
#LIST
fruits=['Apple','Banana','Cherry']
for fruit in fruits:
    print(fruit)
```

```python
#RANGE
for i in range(5):
 print(i)
```

```python
#STRING
for char in 'Hello':
    print(char)
```

```python
integers=[1,2,3,4,5]
for numbers in integers:
 print(numbers)
```

```python
for number in integers:
 print('yup!')
```

```python
for yup in integers:
    print(yup+yup)
```

```python
pulao_dict = {
    'name': 'Rohith',
    'weekly intake': 3,
    'favorite pulaos': ['Tomato Pulao', 'Mint Pulao', 'Curd Pulao']
}
```

```python
# Loop through only values
for item in pulao_dict.values():
    print(item)
```

```python
# Loop through keys and values
for key, value in pulao_dict.items():
    print(key, '->', value)
```

```python
pulao_dict = {
    'name': 'Rohith',
    'weekly intake': 4,
    'favorite pulaos': {
        'veg': ['Tomato Pulao', 'Mint Pulao', 'Curd Pulao'],
        'non-veg': ['Chicken Pulao', 'Mutton Pulao']
    }
}
```

```python
# Loop through top-level keys and values
for key, value in pulao_dict.items():
    print(key, '->', value)
```

```python
# Loop specifically through favorite pulaos
for category, pulaos in pulao_dict['favorite pulaos'].items():
    print(category, "pulaos:", pulaos)
```
